# Train Test Split

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.3,random_state=42)
```

```python
test
```

```python
train.corr()
```

```python
train_x=train.drop('CreditCard',axis=1)
train_x=train_x.drop('ID',axis=1)
train_x=train_x.drop('ZIP Code',axis=1)
test_x=test.drop('CreditCard',axis=1)
test_x=test_x.drop('ID',axis=1)
test_x=test_x.drop('ZIP Code',axis=1)
```

```python
train_y=train['CreditCard']
```

```python
test_x
```

# Logistic Regression

```python
clf=LogisticRegression(random_state=0)
clf.fit(train_x,train_y)
clf.score(train_x,train_y)
```

```python

test_y=test['CreditCard']
test_x.shape
```

```python
clf.score(test_x,test_y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
cross_val_score(clf,test_x,test_y)
```
